```python
import requests
from bs4 import BeautifulSoup

url        = 'https://nomads.ncep.noaa.gov/cgi-bin/filter_gfs_0p25.pl'

main_page  = [a['href'] for a in BeautifulSoup(requests.get(url).text,'html.parser').findAll('a')]

time_page  = [c['href'] for b in [BeautifulSoup(requests.get(a).text,'html.parser').findAll('a') for a in main_page] for c in b if c['href'][-2:] in ('00','12')]

field_page = [c['href'] for b in [BeautifulSoup(requests.get(a).text,'html.parser').findAll('a') for a in time_page] for c in b]

level_page = [BeautifulSoup(requests.get(a).text,'html.parser') for a in field_page]

level_page
```
